```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
from lxml import html
import datetime
import pandas as pd
df = pd.DataFrame(columns=('id', 'name', 'org', 'okreg', 'glosow', 'doswiadczenie', 'klub', 'wiek', 'urodzony', 'wyksztalcenie', 'uczelnie','zawod'))
now = datetime.datetime.now()
```

```python
url = 'https://www.sejm.gov.pl/sejm9.nsf/posel.xsp?id=001'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
osoba = soup.find(id='title_content')
nazwa = osoba.select("div h1")
napis = str(nazwa)[ 5 : -1]
napis = napis[ 0 : -5]
print(napis)
```

```python
partia = soup.find(class_='partia')
dane1r = partia.findAll('p', class_='right')
# print(dane1r[0].get_text())
print(dane1r[1].get_text())
print(dane1r[2].get_text())
print(dane1r[3].get_text())
print(dane1r[4].get_text())
print(dane1r[5].get_text())
```

```python
partia = soup.find(class_='cv')
dane1cv = partia.findAll('p', class_='right')
latka = int(now.year) - int(dane1cv[0].get_text()[ 6 : 10 ])
print(latka)
print(dane1cv[0].get_text())
print(dane1cv[1].get_text())
print(dane1cv[2].get_text())
print(dane1cv[3].get_text())
```

```python
idp = 1
name = napis
org1 = dane1r[1].get_text()
okreg2 = dane1r[2].get_text()
glosow3 = dane1r[3].get_text()
doswiadczenie4 = dane1r[4].get_text()
klub5 = dane1r[5].get_text()
wiek6 = latka
urodzony7 = dane1cv[0].get_text()
wyksztalcenie8 = dane1cv[1].get_text()
uczelnie9 = dane1cv[2].get_text()
zawod10 = dane1cv[3].get_text()
df.loc[idp] = [idp,name,org1,okreg2,glosow3,doswiadczenie4,klub5,wiek6,urodzony7,wyksztalcenie8,uczelnie9,zawod10]
time.sleep(1)
df.head()
```
